# Group 1 — DDSP Timbre Transfer Inference

Load a fine-tuned DDSP checkpoint, run timbre transfer on a session WAV,
and resample the output from 16 kHz to 44.1 kHz.

**Prerequisites:**
- A fine-tuned DDSP checkpoint (from `group1_ddsp_finetune.ipynb`).
- A session WAV file to process.

**Outputs:** A timbre-cloned WAV file at 44.1 kHz / 16-bit PCM.

```python
import sys
from pathlib import Path

# Add notebooks/ to path so we can import from shared/
sys.path.insert(0, str(Path("..").resolve()))

import librosa
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd

from shared.ddsp import OUTPUT_SAMPLE_RATE, run_inference

# --------------- Configuration (edit these) ---------------
CHECKPOINT_DIR = "../../calibrations/finetune_output"  # from finetune notebook
INPUT_WAV = "../../recordings/your_recording.wav"  # session to process
OUTPUT_WAV = "../../output/cloned_output.wav"
```

```python
# ---- Run inference ----

output_path = run_inference(CHECKPOINT_DIR, INPUT_WAV, OUTPUT_WAV)
```

```python
# ---- Playback + comparison ----

print("Input audio:")
ipd.display(ipd.Audio(INPUT_WAV))

print("\nTimbre-cloned output:")
ipd.display(ipd.Audio(OUTPUT_WAV))

# Waveform comparison
input_audio, input_sr = librosa.load(INPUT_WAV, sr=OUTPUT_SAMPLE_RATE, mono=True)
output_audio, output_sr = librosa.load(OUTPUT_WAV, sr=OUTPUT_SAMPLE_RATE, mono=True)

fig, axes = plt.subplots(2, 1, figsize=(14, 5), sharex=True)

t_in = np.arange(len(input_audio)) / input_sr
axes[0].plot(t_in, input_audio, linewidth=0.3)
axes[0].set_ylabel("Amplitude")
axes[0].set_title("Input")

t_out = np.arange(len(output_audio)) / output_sr
axes[1].plot(t_out, output_audio, linewidth=0.3, color="orange")
axes[1].set_ylabel("Amplitude")
axes[1].set_xlabel("Time (s)")
axes[1].set_title("Timbre-Cloned Output")

plt.tight_layout()
plt.show()
```
